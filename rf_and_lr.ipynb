```python
from sklearn.linear_model import LogisticRegression
```

```python
x_column = dt.columns.to_list()[:-1]
xt = dt[x_column]
scaler = StandardScaler()
x = scaler.fit_transform(xt)
```

```python
y = dt['species']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=.2, random_state=0)
```

```python
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
ypred = model.predict(x)
```

```python
confusion_matrix(y, ypred)
```

```python
print(classification_report(y, ypred))
```

```python
from sklearn.impute import SimpleImputer
```

```python
impute = SimpleImputer()
```

```python
column = ['R&D Spend', 'Marketing Spend']
```

```python
df[column] = impute.fit_transform(df[column]).astype(int)
```

```python
df
```

```python
label = LabelEncoder()
df['State'] = label.fit_transform(df['State'])
```

```python
df
```

```python
x_df = df.columns.to_list()[:-1]
y = df['Profit']
```

```python
xdf = df[x_df]
xd = scaler.fit_transform(xdf)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(xd, y, test_size=.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(xtrain, ytrain)
```

```python
rf.score(xd, y) * 100
```

```python
ypred_2 = rf.predict(xd)
```

```python
df['prediction'] = ypred_2
```

```python
df
```

```python
ax = df['Profit'].plot(kind='kde', figsize=(15,5))
df['prediction'].plot(kind='kde', figsize=(15,5), ax=ax)
plt.legend()
plt.show()
```
