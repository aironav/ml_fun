```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.preprocessing import StandardScaler
import os
```

```python
dt = pd.read_csv('../dataset/sample_data.csv')
```

```python
dt
```

```python
imputer = SimpleImputer()
```

```python
column = ['salary', 'age']
```

```python
dt[column] = imputer.fit_transform(dt[column]).astype(int)
```

```python
dt
```

```python
dt.nunique()
```

```python
encoder = LabelEncoder()
dt['happy'] = encoder.fit_transform(dt['happy'])
```

```python
dt
```

```python
one_hot_encoder = OneHotEncoder(drop = 'first')
country_enc = one_hot_encoder.fit_transform(dt[['country']]).toarray()
```

```python
country_enc
```

```python
new_dt = pd.DataFrame(country_enc, columns=['italy','spain'])
```

```python
dt = pd.concat([new_dt, dt], axis=1)
```

```python
dt.drop(columns=['country'], axis=1, inplace=True)
```

```python
dt
```

```python
colu = dt.columns.tolist()[:-1]
```

```python
colu
```

```python
scaling = StandardScaler()
dt[colu] = scaling.fit_transform(dt[colu])
```
